```python
# in json

import json



bookfile='books.json'


"""
json
[
   {
     'name': name
     'author':author
     'read': True / False
   
   },
   
   {
   .......
   .......
   
   }
]

"""

def create_book():
    with open(bookfile,'a') as file:
        json.dump([],file)
        

    
    
def book_list():
    with open(bookfile,'r') as file:
        return json.load(file)
    
    
def _save2file(books):     
    with open(bookfile,'w') as file:
        json.dump(books,file)

def add_book():
    name=input('Enter the name of the book : ')
    author=input('Enter the name of the author : ')
    books=book_list()
    books.append({'name':name,'author':author,'read':False})
    _save2file(books)
        

        
    
def mark_read():
    name=input('Enter tha name of the book you want to mark as read : ')
    books=book_list()
    for book in books:
        if book['name']==name:
            book['read']=True
            
    _save2file(books)
    
    

        
def delete_book():
    name=input('Enter the name of the book you want to delete : ')
    books=book_list()      #list of dicitonaries
    new_book=[book for book in books if book['name'] != name]
    _save2file(new_book)
    
    
def menu():
    create_book()
    sel = input("Enter 'a' to Add a book,"
                "'s' to see the list of added books,"
               "'d' to delete a book,"
               "'m' to mark a book as read,"
               "'q' to quit! : ")
    
    while sel!='q':
        if sel=='a':
            add_book()
        elif sel=='s':
            print(book_list())
            
        elif sel=='d':
            delete_book()
        elif sel=='m':
            mark_read()
            
        else:
            print('UNKNOWN COMMAND!')
            
        sel = input("Enter 'a' to Add a book,"
                          "'s' to see the list of added books,"
                          "'d' to delete a book,"
                          "'m' to mark a book as read,"
                          "'q' to quit! : ")
        
        
menu()
```
