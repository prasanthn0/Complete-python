```python
bookfile='books.txt'


def add_book():
    name=input('Enter the name of the book : ')
    author=input('Enter the name of the author : ')
    with open(bookfile,'a') as file:
        file.write(f'{name},{author},0\n')
        

def book_list():
    with open(bookfile,'r') as file:
        lines=[line.strip().split(',') for line in file.readlines()]
        
    return [
        
        {'name':line[0],'author':line[1],'read':line[2]}
        
        for line in lines
    ]    #List of dictionaries
 
  

    
def mark_read():
    name=input('Enter tha name of the book you want to mark as read : ')
    books=book_list()
    for book in books:
        if book['name']==name:
            book['read']='1'
            
    _save2file(books)
    
    
def _save2file(books):     #list of dicitonaries and each dictionary is a book
    with open(bookfile,'w') as file:
        for book in books:
            file.write(f"{book['name']},{book['author']},{book['read']}\n")
        
def delete_book():
    name=input('Enter the name of the book you want to delete : ')
    books=book_list()      #list of dicitonaries
    new_book=[book for book in books if book['name'] != name]
    _save2file(new_book)
    
    
def menu():
    sel = input("Enter 'a' to Add a book,"
                "'s' to see the list of added books,"
               "'d' to delete a book,"
               "'m' to mark a book as read,"
               "'q' to quit! : ")
    
    while sel!='q':
        if sel=='a':
            add_book()
        elif sel=='s':
            print(book_list())
            
        elif sel=='d':
            delete_book()
        elif sel=='m':
            mark_read()
            
        else:
            print('UNKNOWN COMMAND!')
            
        sel = input("Enter 'a' to Add a book,"
                          "'s' to see the list of added books,"
                          "'d' to delete a book,"
                          "'m' to mark a book as read,"
                          "'q' to quit! : ")
        
        
menu()
```
