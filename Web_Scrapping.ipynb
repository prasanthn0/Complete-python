```python
#locator.py





class PageLocators:
    QUOTE='div.quote'
```

```python
#quote locators

class QuoteLocator:
    AUTHOR='small.author'
    CONTENT='span.text'
    TAGS='div.tags a.tag'
    
```

```python
#quote parsers

class QuoteParser:
    def __init__(self,parent):
        self.parent=parent
    
    
    @property
    def content(self):
        locator=QuoteLocator.CONTENT
        return self.parent.select_one(locator).string
    
    @property
    def author(self):
        locator=QuoteLocator.AUTHOR
        return self.parent.select_one(locator).string
    
    @property
    def tags(self):
        locator=QuoteLocator.TAGS
        return [e.string for e in self.parent.select_one(locator)]
    
    def __repr__(self):
        return f'<QUOTE {self.content}> By <{self.author}>'
        
```

```python
#import pagelocator and quote parser

from bs4 import BeautifulSoup

class QuotesPage:
    def __init__(self,page):
        self.soup=BeautifulSoup(page,'html.parser')
        
    @property
    def quotes(self):
        locator=PageLocators.QUOTE
        quote_tags=self.soup.select(locator)
        return [QuoteParser(e) for e in quote_tags]
```

```python
#locators : #all-books-page #book-locators


class BookLocators:
    NAME_LOCATOR= 'article.product_pod h3 a'
    LINK_LOCATOR = 'article.product_pod h3 a'
    PRICE_LOCATOR='article.product_pod p.price_color'
    RATING_LOCATOR = 'article.product_pod p.star-rating'
    
class AllBooksPageLocator:
    BOOKS='div.page_inner section li.col-xs-6.col-sm-4'
    PAGER='div.page_inner section ul.pager li.current'
    
```

```python
#pages 

from bs4 import BeautifulSoup
import re

class AllBooksPage:
    
    def __init__(self,page):
        self.soup=BeautifulSoup(page,'html.parser')

    @property
    def books(self):
        # return self.soup.select(AllBooksPageLocator.BOOKS) will give us tags 
        return [BookParser(e) for e in self.soup.select(AllBooksPageLocator.BOOKS)]
    
    @property
    def page_count(self):
        content=self.soup.select_one(AllBooksPageLocator.PAGER).string
        pattern= 'Page [0-9]+ of ([0-9]+)'
        matcher=re.search(pattern,content)
        return int(matcher.group(1))
        
    
```

```python
#parser

import re

from bs4 import BeautifulSoup

class BookParser:
    
    
    
    
    #lives inside the class and not inside self
    RATINGS={
        'one': 1,
        'Two':2,
        'Three':3,
        'Four':4,
        'Five':5
    }
    
    
    def __init__(self,parent):
        self.parent=parent
        
    def __repr__(self):
        return f'<Book {self.name} , {self.price}Euro and {self.rating} stars>'
    
    @property   # instead of using this as a function call self.name() , can do self.name
    def name(self):
        #locator='article h3 a'  But we can be even more specific with the class name
        locator=BookLocators.NAME_LOCATOR  #CSS locator
        item_link= self.parent.select_one(locator)
        item_name=item_link.attrs['title']
        return item_name


    @property
    def price(self):
        locator=BookLocators.PRICE_LOCATOR
        price=self.parent.select_one(locator).string    #£51.77
        pattern= '£([0-9+\.[0-9]+)'
        matcher=re.search(pattern,price)
        #print(matcher.group(0))
        #print(matcher.group(1))
        #print(type(matcher.group(1)))
        return float(matcher.group(1))


    
    @property
    def rating(self):
        locator=BookLocators.RATING_LOCATOR
        starRatingTag=self.parent.select_one(locator)
        classes=starRatingTag.attrs['class']  #['star-rating','Three']
        rating_classes=[r for r in classes if r!='star-rating']
        #rating= filter(lambdax:x!='star-rating',classes)
        rating=BookParser.RATINGS.get(rating_classes[0], 0)   #will return 0 if rating property is not found
        return rating
    
    
    @property
    def link(self):
        locator=BookLocators.LINK_LOCATOR
        item_link=self.parent.select_one(locator)
        return item_link.attrs['href']
    
    


    
```
